```python
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
models = {
    "RFR" : RandomForestRegressor(),
    "XGB" : XGBRegressor(),
    "Cat" : CatBoostRegressor(),
    "GBR": GradientBoostingRegressor()
}
```

# **To predict Count:-**

```python
df.drop(['registered','casual','atemp'],inplace=True,axis=1)
```

```python
df
```

```python
from sklearn.metrics import mean_absolute_error,root_mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('cnt',axis=1)
y = df['cnt']

x_train,x_temp,y_train,y_temp = train_test_split(x,y,test_size=0.2)
x_val,x_test,y_val,y_test = train_test_split(x_temp,y_temp,test_size=0.5)

results = {}

for name,model in models.items():
    model.fit(x_train,y_train)
    preds = model.predict(x_val)
    mae = mean_absolute_error(y_val,preds)
    rmse = root_mean_squared_error(y_val,preds)
    r2 = r2_score(y_val,preds)
    results[name] = {"MAE":mae,"RMSE":rmse,"R2":r2}

results_df = pd.DataFrame(results).T
print(results_df)
```

```python
df
```

```python
top_model = CatBoostRegressor()
```

# **To check for overfitting:-**

```python
top_model.fit(x_train,y_train)
preds = top_model.predict(x_test)
r2 = r2_score(y_test,preds)
```

```python
r2
```

```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(top_model, x, y, cv=5, scoring="r2")
print("CV R2 scores:", cv_scores)
print("Mean CV R2:", cv_scores.mean())
```

## **To try to overcome overfitting:-**

```python
cols = ['season','mnth','holiday','weekday','workingday','weathersit']
for col in cols:
    df[col] = df[col].astype('category')
```

```python
df.info()
```

```python
from sklearn.model_selection import RandomizedSearchCV

param_dist = {
    'depth': [4, 6, 8, 10],
    'learning_rate': np.linspace(0.03, 0.2, 8),
    'l2_leaf_reg': [1, 3, 5, 7, 10],
    'bagging_temperature': [0, 0.25, 0.5, 0.75, 1.0],
    'random_strength': [0, 0.5, 1.0, 1.5, 2.0],
    'iterations': [600, 900, 1200, 1500, 2000],
    'subsample': [0.6, 0.8, 1.0]
}

rs_model = RandomizedSearchCV(
    estimator=top_model,
    param_distributions=param_dist,
    n_iter=5,
    scoring='r2',
    cv=3,
    n_jobs=-1,
    refit=True
)
rs_model.fit(x_train, y_train)   # pass cat_features here
best_cat_model = rs_model.best_estimator_
```

```python
best_cat_model
```

```python
def best_model_evaluation(model):
    model.fit(x_train,y_train)
    preds = model.predict(x_val)
    mae = mean_absolute_error(y_val,preds)
    rmse = root_mean_squared_error(y_val,preds)
    r2 = r2_score(y_val,preds)
    print("MAE = ",mae,"\n","RMSE = ",rmse,"\n","R2 = ",r2)
```

```python
best_model_evaluation(best_cat_model)
```

```python
preds = best_cat_model.predict(x_test)
mae = mean_absolute_error(y_test,preds)
rmse = root_mean_squared_error(y_test,preds)
r2 = r2_score(y_test,preds)
print("MAE = ",mae,"\n","RMSE = ",rmse,"\n","R2 = ",r2)
```
